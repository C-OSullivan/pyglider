```python
import scipy.stats as stats

with xr.open_dataset('L1-timeseries/rosie713-20180915T1202_L1.nc', decode_times=False) as ds:

    profiles = np.unique(ds.profile_index)
    profiles = [p for p in profiles if (~np.isnan(p) and not (p % 1)
                                                and (p > 0))]
    
    profile_bins = np.arange(0.5, max(profiles)+0.51, 1.)
    
    Nprofiles = len(profiles)
    depthbins = np.arange(0, 1100.1)
    depths = depthbins[:-1] + 0.5

    dsout = xr.Dataset(coords={'depth': ('depth', depths), 
                               'profile': ('time', profiles)},
                      )
    for td in ('time', 'longitude', 'latitude'):
        dat = np.zeros(len(profiles))
        for n, p in enumerate(profiles):
            ind = np.where(ds.profile_index == p)
            dat[n] = ds[td][ind].values.mean()
        dsout[td] = (('time'), dat, ds[td].attrs )

    print('temperature' in ds)
    for k in ds.keys():
        if k not in ['time', 'longitude', 'latitude', 'depth']:
            dat = np.zeros((len(depths), len(profiles))) * np.NaN
            good = np.where(~np.isnan(ds[k]))[0]
            print(len(good))
            print(k)
            dat, xedges, yedges, binnumber = stats.binned_statistic_2d(
                    ds['profile_index'].values, ds['depth'].values, values=ds[k].values, statistic='mean', 
                    bins=[profile_bins, depthbins])
            print(np.shape(dat), len(profiles))
            print('done', k)
            dsout[k] = (('depth', 'time'), dat.T, ds[td].attrs )
            
            print(k)
            if 0:
                if len(good) > 0:
                    for n, p in enumerate(profiles):
                        ind = np.where(ds.profile_index[good] == p)[0]
                        dd = ds[k].isel(time=good[ind])
                        dsout[k][:, n] = dd.groupby_bins('depth', depthbins).mean()
        
        
```
